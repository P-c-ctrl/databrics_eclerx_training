```python
csv_path ="dbfs:/Volumes/workspace/test_schema/circuit/circuits.csv"

df = spark.read.csv( csv_path
                    ,header=True
                    ,inferSchema=True
                    )
```

```python
df.display ()
```

```python
from pyspark.sql.functions import current_date

df_final = df.withColumnsRenamed ({"circuitId": "circuit_id", "circuitRef": "circuit_ref"})\
                .withColumn ("ingestion_date", current_date() )\
                .drop("url")
```

```python
%sql
create schema if not exists formula1
```

```python
df_final.write.mode("overwrite").saveAsTable ("formula1.circuits")
```

```python
%sql
select * from formula1.circuits
```
