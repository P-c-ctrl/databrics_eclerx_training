```sql
select mnth, round(max(temp), 2) as maxtemp from bikesharing group by mnth 
order by maxtemp desc
```

```sql
create or replace view max_month as 
select mnth, round(max(temp), 2) as maxtemp from bikesharing group by mnth 
order by maxtemp desc
```

```sql
select * from max_month
```

```sql
desc extended max_month
```

```sql
desc extended bikesharing
```

```sql
create or replace temp view total_holidays as
select yr, season, sum(holiday) as total_holidays from bikesharing group by yr, season
```
